```python
import numpy as np
import matplotlib.pyplot as plt

rpm = np.linspace(0, 5000, 100)

T_max = 61   # example Nm
base_rpm = 600
max_rpm = 6000

torque = []

for r in rpm:
    if r <= base_rpm:
        torque.append(T_max)
    else:
        torque.append(T_max * (1 - (r - base_rpm)/(max_rpm - base_rpm)))

plt.plot(rpm, torque)
plt.xlabel("RPM")
plt.ylabel("Torque (Nm)")
plt.title("Torque-RPM Curve")
plt.show()
```
